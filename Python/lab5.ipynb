```python
class Person:
    def __init__(self, name, money, mood="Happy", healthRate="100%"):
        self.name = name
        self.money = money
        self.mood = mood
        self.healthRate = healthRate
        print("Person Constructor")

    def sleep(self, hours):
        if hours > 7:
            self.mood = "Lazy"
        elif hours < 7:
            self.mood = "Tired"
        else:
            self.mood = "Happy"

    def eat(self, meals):
        if meals == 3:
            self.healthRate = "100%"
        elif meals == 2:
            self.healthRate = "75%"
        elif meals == 1:
            self.healthRate = "50%"

    def buy(self, items):
        self.money -= items * 10
```

```python
class Employee(Person):
    def __init__(
        self, id, name, money, mood, healthRate, car, email, salary, distanceToWork
    ):
        Person.__init__(self,name, money, mood, healthRate)
        self.id = id
        self.car = car
        self.email = email
        self.salary = salary
        self.distanceToWork = distanceToWork
        print("Employee Constructor")

    def work(self, hours):
        if hours > 8:
            self.mood = "Tired"
        elif hours < 8:
            self.mood = "Lazy"
        else:
            self.mood = "Happy"

    def drive(self, distance, velocity):
        self.car.run(distance, velocity)

    def refuel(self, gasAmount=100):
        if gasAmount > 0:
            self.car.fuelRate += gasAmount
            if self.car.fuelRate > 100:
                self.car.fuelRate = 100
        else:
            print("Can't ReFull")

    def send_mail(self):
        print("Send Mail")
```

```python
class Office:
    employeesNum = 0

    def __init__(self, name, employees):
        self.name = name
        self.employees = employees
        numsEmp = len(employees)
        Office.changeNumEmployees(numsEmp)
        print("Office Constructor")

    def get_all_employees(self):
        return self.employees

    def get_employee(self, empId):
        for employee in self.employees:
            if employee.id == empId:
                return employee
        return -1

    def hire(self, employee):
        self.employees.append(employee)
        Office.employeesNum += 1

    def fire(self, empId):
        for employee in self.employees:
            if employee.id == empId:
                self.employees.remove(employee)
                Office.employeesNum -= 1

    def reward(self, empId, reward):
        for employee in self.employees:
            if employee.id == empId:
                employee.money += reward

    def deduct(self, empId, deduction):
        for employee in self.employees:
            if employee.id == empId:
                employee.money -= deduction

    def check_lateness(self, empId, moveHour):
        if moveHour == 0:
            self.reward(empId, 10)
        else:
            self.deduct(empId, 10)

    @classmethod
    def changeNumEmployees(cls,nums):
        Office.employeesNum = nums

    @staticmethod
    def calculate_lateness(targetHour, moveHour, distance, velocity):
        time_taken = distance / velocity
        expected_arrival = targetHour - time_taken
        return moveHour > expected_arrival
```

```python
class Car:
    def __init__(self, name, fuelRate, velocity):
        self.name = name
        self.fuelRate = fuelRate
        self.velocity = velocity
        print("Car Constructor")

    def run(self, distance, velocity):
        if velocity > 0 and velocity <= 200:
            self.velocity = velocity
            if self.fuelRate > 0:
                fuelConsumed = (distance / 10) * (self.fuelRate * 0.1)
                if fuelConsumed > self.fuelRate:
                    traveledDistance = (self.fuelRate / (self.fuelRate * 0.1)) * 10
                    remainingDistance = distance - traveledDistance
                    self.stop(remainingDistance)
                else:
                    self.stop(0)
                    self.fuelRate -= fuelConsumed
            else:
                print("No fuel, Please Refuel")
        else:
            print("Car Velocity Is out of Range")

    def stop(self, remainingDistance):
        self.velocity = 0
        if remainingDistance == 0:
            print("You arrive the Destination")
        else:
            print(f"Oops,You have not reached your destination [Remaining Distance : {self.run.remained_distance} km]")
```
